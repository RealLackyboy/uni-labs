```python
#МЕТОД  ХОРД 
e = 0.0001 
a0 = -2 
b0 = 4 
a = a0
b = a0 + 1
a_ = a
b_ = b

x = a_ 
x_copy = x

def fx(x):  
    res = 3 * x**3 - 9 * x**2 - 12 * x +3 
    return res 
 
def rozr(): 
    x = round(a_ - (fx(a_)) * ((b_ - a_)/(fx(b_) - fx(a_))), 7)
    global x_copy
    x_copy = x

while b <= b0:
    if fx(a_) * fx(b_) < 0: 
        rozr()

        while abs(3 * x**3 - 9 * x**2 - 12 * x +3) > e and abs(fx(x_copy)-x_copy) > e: 
            
            if round(fx(x_copy), 4) == 0:
                print(f'корінь = {round(x_copy, 5)}')
                break
            
            rozr() 
            b_ = x_copy


    a += 1
    b += 1 
    a_ = a
    b_ = b
    
```
